# Espncrickinfo scorecard scrapper

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.espncricinfo.com/series/big-bash-league-2021-22-1269637/match-schedule-fixtures-and-results')

html_data = page.content

soup = BeautifulSoup(html_data)

match_page_list = []
for a in soup.find_all('a', href=True):
  if 'full-scorecard' in a['href']:
    match_page_list.append(a['href'])
print(match_page_list)
print(len(match_page_list))
```

```python
emt_bat_df = pd.DataFrame()
emt_ball_df = pd.DataFrame()

for subUrl in match_page_list:
  try:
    t1_batting = pd.read_html(f"https://www.espncricinfo.com/{subUrl}")[0]
    t2_bowling = pd.read_html(f"https://www.espncricinfo.com/{subUrl}")[1]
    t2_batting = pd.read_html(f"https://www.espncricinfo.com/{subUrl}")[2]
    t1_bowling = pd.read_html(f"https://www.espncricinfo.com/{subUrl}")[3]

    merg_batting = pd.concat([t1_batting, t2_batting],ignore_index=True)
    merg_batting = merg_batting[merg_batting["BATTING"].str.contains("Did not bat") == False]
    merg_batting = merg_batting[merg_batting["BATTING"].str.contains("Fall of wickets:") == False]
    merg_batting = merg_batting[merg_batting["BATTING"].str.contains("Extras") == False]
    merg_batting = merg_batting[merg_batting["BATTING"].str.contains("TOTAL") == False]
    merg_batting = merg_batting.dropna(axis=1, how='all')
    emt_bat_df = pd.concat([emt_bat_df, merg_batting],ignore_index=True)

    merg_bowling = pd.concat([t1_bowling, t2_bowling],ignore_index=True)
    merg_bowling = merg_bowling[~merg_bowling.M.str.contains(" ", na=False)]
    emt_ball_df = pd.concat([emt_ball_df, merg_bowling],ignore_index=True)
  except:
    print(subUrl)
    

final_data = pd.concat([emt_bat_df, emt_ball_df], axis=1,ignore_index=True)
print(final_data)
```

```python
final_data.to_csv('bblData.csv')
```

# Player Historic data

```python
page = requests.get('https://www.espncricinfo.com/series/big-bash-league-2022-23-1324623/hobart-hurricanes-vs-melbourne-renegades-14th-match-1324637/match-squads')

html_data = page.content

soup = BeautifulSoup(html_data)

palyer_list_url = []
for a in soup.find_all('a', href=True):
  if '/player/' in a['href']:
    palyer_list_url.append(a['href'])
print(palyer_list_url)
print(len(palyer_list_url))
```

```python
combiled_player_data = pd.DataFrame()
player_token_list = []
for player_name in palyer_list_url:
  try:
    playes_data = pd.read_html(f"https://www.espncricinfo.com{player_name}/matches")[0]
    playes_data.drop(['Match','Format','Date'], axis=1,inplace = True)
    num_list = list(range(len(playes_data)))
    playes_data['Match']=pd.Series(sorted(num_list, reverse=True))
    playes_data['Name'] = player_name.split('/')[-1]
    if 'Bowl' not in playes_data:
      playes_data['Bowl']=0

    playes_data = playes_data.reindex(sorted(playes_data.columns), axis=1)
    playes_data['Bat']=playes_data['Bat'].replace('\*','',regex=True)
    try:
      playes_data['Bat'] = playes_data['Bat'].str.split(' ').str[0]
    except:
      pass
    playes_data['Bat']=playes_data['Bat'].replace('\--','0',regex=True).astype(int)
    try:
      playes_data['Bowl'] = playes_data['Bowl'].str.split('/').str[0]
    except:
      pass
    playes_data['Bowl']=playes_data['Bowl'].replace('\--','0',regex=True).astype(int)
    combiled_player_data = pd.concat([combiled_player_data, playes_data],ignore_index=True)
    print(player_name.split('/')[-1])
    player_token_list.append(player_name.split('/')[-1])
  except:
    print(f"------------- PASSED {player_name.split('/')[-1]} -----------")
```
